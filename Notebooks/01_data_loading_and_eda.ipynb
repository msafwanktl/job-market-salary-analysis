Combining Dictionary + Condition

```python
job_role = "Data Analyst"
salary = 48000

if salary >= 50000:
    print(job_role, "is well paid")
else:
    print(job_role, "is entry level")
```

```python
person = {
    "name": "Your Name",
    "age": 23
}

if person["age"] >= 18:
    print("Adult")
else:
    print("Minor")
```

Loop + Condition

```python
jobs = [
    {"title": "Data Analyst", "salary": 48000},
    {"title": "Data Scientist", "salary": 70000},
    {"title": "Business Analyst", "salary": 52000}
]

for job in jobs:
    if job["salary"] >= 50000:
        print(job["title"], "→ Good salary")
    else:
        print(job["title"], "→ Entry level")
```
