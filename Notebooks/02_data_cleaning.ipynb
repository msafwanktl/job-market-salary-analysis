Loading Dataset

```python
import pandas as pd

df = pd.read_csv("/ds_salaries.csv")
df.head()
```

Remove Unnecessary Column

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df.columns
```

Check for Missing Values

```python
df.isnull().sum()
```

**Understand Categorical Columns**

```python
df.select_dtypes(include="object").columns
```

Clean experience_level

```python
experience_map = {
    "EN": "Entry",
    "MI": "Mid",
    "SE": "Senior",
    "EX": "Executive"
}

df["experience_level"] = df["experience_level"].map(experience_map)
```

```python
df["experience_level"].value_counts()
```

Clean employment_type

```python
employment_map = {
    "FT": "Full-time",
    "PT": "Part-time",
    "CT": "Contract",
    "FL": "Freelance"
}

df["employment_type"] = df["employment_type"].map(employment_map)
```

Convert work_year to Category

```python
df["work_year"] = df["work_year"].astype(str)
```
